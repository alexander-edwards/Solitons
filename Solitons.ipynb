```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
```

1. Prep work

$$ u(x, t)=-\frac{c}{2} \operatorname{sech}^{2}\left[\frac{1}{2} \sqrt{c}\left(x-c t-\xi_{0}\right)\right] $$

```python
def f(x, t): 
    c = 0.1
    z = 1
    return -c / 2 * (1/np.cosh(1/2 * c**0.5 * (x - c*t - z)))**2


def visualisation_3d(): 
    x = np.linspace(-10, 20, 100)
    t = np.linspace(-10, 20, 100) 
    X, T = np.meshgrid(x, t) 
    vf = np.vectorize(f)
    z = vf(X, T) 
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1, projection='3d')
    ax.plot_surface(X, T, z, cmap = 'viridis')
    plt.gca().set_aspect('equal', adjustable='box')
    
visualisation_3d()    

    
```
